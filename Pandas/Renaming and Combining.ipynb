```python
import pandas as pd

df = pd.read_csv("Flavors.csv")
```

### Renaming Column Name

```python
df = df.rename(columns={"Total Rating":"Rating"})
```

### rename() lets you rename index or column values by specifying a index or column keyword parameter, respectively.
### Here is an example using it to rename some elements of the index.

```python
df = df.rename(index={0:"1st",1 : "2nd"})
```

#### rename_axis() is used to be specipy the name of axis what are the values in axis "X"  and what are the values in axis "Y"

```python
df = df.rename_axis("Chocolate_List" , axis="rows")
df = df.rename_axis("fields", axis="columns")
df
```
