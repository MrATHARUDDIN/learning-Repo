```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date = pd.date_range(start="2024-10-15", end="2025-04-28")

views = 100 + np.cumsum(np.random.normal(loc=0.5, scale=5, size=len(date)))

df = pd.DataFrame({
    "Date": date,
    "Daily_Views": views
})

# Calculate 21-day moving average
df["21D_Ma"] = df["Daily_Views"].rolling(window=21).mean()


plt.figure(figsize=(12,5)) # figure of the plot
plt.title("Youtube view")
plt.xlabel("Date")
plt.ylabel("View")
plt.plot(df["Date"],df["Daily_Views"])

plt.plot(df["Date"],df["21D_Ma"]) # Moving Average
plt.show()
```
