```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdate

# read csv file
df = pd.read_csv("apple.csv")

# Convert "Date" column to datetime
df["Date"] = pd.to_datetime(df["Date"])

plt.figure(figsize=(10, 5))
plt.plot(df["Date"],df["AAPL_Close"])

plt.gca().xaxis.set_major_formatter(mdate.DateFormatter("%Y"))
plt.gca().xaxis.set_major_locator(mdate.YearLocator())
# plt.gca() --> get current Axes
# .xaxis.set_major_formatter(...) --> sets the format of the major ticks on the x-axis
# mdate.DateFormatter("%Y") -->  Formats the date to display only the year

# .xaxis.set_major_locator(...) --> Contols where the major ticks appear.
# mdate.YearLocator() --> It tell the metaplotlib to place ticks once per year


# plot title
plt.title("Apple Stock Price (Monthly)")
plt.xlabel("Year")
plt.ylabel("Price")

plt.show()
```
