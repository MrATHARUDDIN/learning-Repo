```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name': ['Alice', 'Bob', 'Charlie',"rose"],
    'Age': [35, 30, 25, 40],
    'City': ['NY', 'LA', 'UK', "Jupiter"]
}
df = pd.DataFrame(data)

print(df)
```

```python
# 7.1 Multi-Indexing
df_multi = df.set_index(['City', 'Name'])
print(df_multi)

df.plot()
plt.show()
```
